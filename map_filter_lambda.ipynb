```python
lst=[]

#prime number generating function
def toPrintPrimeNo():
    for i in range(2,10000):
        j=i//2
        for k in range(2,j+1):
            if(i%k)==0:
                break
        else:
            lst.append(i)
    return lst

#To filter
toFilterPrimeNo=filter(lambda x:(x<=2500),toPrintPrimeNo())

#to print filtred data
print(list(toFilterPrimeNo))



def unitDigitAsThree(num):
    if(num%10!=3):
        return num
   

toCheckUnitDigit=map(unitDigitAsThree,toPrintPrimeNo())
print(list(toCheckUnitDigit))
```
